# Project 2 - Regression to algebra multiplication

## Goal
To get first hand experience using tensorflow, after regression to naive algebra function $y = x + c$, I want to regression to algebra mulitiplication.

## Formulate inputs and outputs
Input training data $X$: a list of uniformly distributed tuples $x_0, x_1$, where both $x_0$ and $x_1$ are floats within $[-1.0, 1.0]$.

Output training data $Y$: for each tuple $x_0, x_1$, corresponding $y = x_0 * x_1$

Target function: $y = x_0 * x_1$

Note that in this project, I normalized the input training data $X$ to within $[-1, 1]$ to simplify data preprocessing and forcus on models.

```python
import numpy as np
import tensorflow as tf

x0 = np.random.uniform(-1, 1, 1000)
x1 = np.random.uniform(-1, 1, 1000)
x_train = np.array([x0, x1], dtype=np.float32).T
y_train = np.array([x[0]*x[1] for x in x_train], dtype=np.float32)
print ("X_train shape: ".format(x_train.shape))
print (x_train[0:5])
print ("Y_train shape: ".format(y_train.shape))
print (y_train[0:5])
```

## Try Keras models

It's clear that models with one hidden layer can *not* simulate algebra multiplication. I will start with two hidden-layer models.

```python
# Create a two hidden layer model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, kernel_initializer="random_uniform", activation='relu', input_shape=[2]),
    tf.keras.layers.Dense(1)
]
)

model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mae'])

model.fit(x_train, y_train, epochs=100)
model.evaluate(x_train, y_train)
```

## Summarize the model

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model, model_to_dot
plot_model(model, to_file='model.png')
```

## Test the first keras model

### Sanity check difference bween y_train and y_predict

```python
y_predict = model.predict(x_train)
import matplotlib.pyplot as plt
plt.plot(x0, x0, 'r-')
plt.scatter(y_train, y_predict)
```

This model fits pretty well to algebra multiplication $ y = x_0 * x_1 $ for $x_train$ inputs.

### Create test data with floats within $[-1, 1]$ and check predicted outputs

```python
x_test_0 = np.random.uniform(-1, 1, 100)
x_test_1 = np.random.uniform(-1, 1, 100)
x_test = np.array([x_test_0, x_test_1], dtype=np.float32).T
y_test = np.array([x[0] * x[1] for x in x_test])
y_predict = model.predict(x_test)
plt.plot(x0, x0, 'r-')
plt.scatter(y_test, y_predict)
```

This model fits test data pretty well, when input test floats are all within $[-1, 1]$.

### Create test data out of $[-1, 1]$ (model never seen) and check predicted outputs
This model has never seen inputs smaller than $-1$ nor larger than $1$. I want to see to what extend, this model still regession to algebra mulitiplication

```python
x_test_0 = np.random.uniform(-10, 10, 100)
x_test_1 = np.random.uniform(-10, 10, 100)
x_test = np.array([x_test_0, x_test_1], dtype=np.float32).T
y_test = np.array([x[0] * x[1] for x in x_test])
y_predict = model.predict(x_test)
plt.plot(x_test_0, x_test_0, 'r-')
plt.scatter(y_test, y_predict)
```

It's not supprising that this model do not fit well when input floats are smaller than $-1$ or larger than $1$.

### Lesson 1: Must normalize input training data and testing data so that they are in the same range.
### Lesson 2: Must understand scope of input that a model works, and do not assume a model fit unseen data out of the box.
